# 03. URL化（URLify）
### 問題の概要
* 与えられた文字列内のすべてのスペースを %20 に置き換えるメソッドを作成してください。
* 文字列の末尾には新しい文字を収容するのに十分なスペースがあり、文字列の「真の」長さ（有効な文字数）がわかっているものとします。

解法 1：アルゴリズム重視（双ポインタ法）
この解法は、メモリ効率を重視する低レイヤ言語（CやJavaなど）の考え方に基づいています。Pythonでは文字列が不変（Immutable）であるため、一度リストに変換してから操作を行います。

```python
class Solution:
    def replaceSpaces(self, chars: str, true_length: int) -> str:
        s_list = list(chars)
        space_count = 0
        for i in range(true_length):
            if chars[i] == ' ':
                space_count += 1

        new_index = true_length + space_count *2

        for i in range(true_length - 1, -1, -1):
            if s_list[i] == ' ':
                s_list[new_index - 1] = '0'
                s_list[new_index - 2] = '2'
                s_list[new_index - 3] = '%'
                new_index -= 3
            else:
                s_list[new_index - 1] = s_list[i]
                new_index -= 1
        return "".join(s_list).strip()
```

## 思考プロセスと考察：
なぜ後ろから操作するのか？ 前から置換を始めると、後続の文字を何度も後ろにずらす必要があり、計算量が $O(N^2)$ になってしまいます。後ろから操作することで、各文字の移動は一度だけで済み、計算量は $O(N)$ に抑えられます。メモリ効率： 本来の意図は「追加のメモリを確保せずに既存の配列を書き換える」ことにあります。アルゴリズムの本質を理解していることをアピールするのに最適な解法です。

解法 2：Pythonicなアプローチ（組み込みメソッド）
実際の開発現場で最も使われる、Pythonの強力な組み込み機能を活用したシンプルで高速な方法です。

```python
def urlify_pythonic(s: str, true_length: int) -> str:
    # 1. スライス [ :true_length ] で有効な部分だけを切り出す
    # 2. replace メソッドで一括置換
    return s[:true_length].replace(" ", "%20")
```

## 思考プロセスと考察：
スライスの重要性： 問題文にある「末尾の十分なスペース」は、Pythonの replace を使う際には不要なノイズとなります。そのため、まず s[:true_length] を使って必要な部分だけを抽出するのがポイントです。パフォーマンス： replace メソッドは内部的に C言語 で実装・最適化されているため、自身で for ループを書くよりも実行速度が非常に速いです。計算量： 時間計算量は $O(N)$、空間計算量は新しい文字列を生成するため $O(N+M)$ となります。
