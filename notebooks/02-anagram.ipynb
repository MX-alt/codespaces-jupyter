# 02. アナグラム判定

### 問題
2つの文字列 `s1` と `s2` が与えられたとき、一方を並べ替えてもう一方にできるか（アナグラムか）を判定してください。

**条件**:
- すべて英小文字とする。
- 実行速度を重視し、ネスト（入れ子）の少ない読みやすいコードを目指す。

# 解法 1: ソート法 (Sorting)

```python
def is_anagram_sort(s1: str, s2: str) -> bool:
    if len(s1) != len(s2):
        return False
    # 文字列をソートして比較するだけの簡潔な方法
    return sorted(s1) == sorted(s2)
```

# 解法 2: カウント法 (Counting / Frequency Array)

```python
def CheckPermutation(s1: str, s2: str) -> bool:
    if len(s1) != len(s2):
        return False
        
    counts = [0] * 26

    for char in s1:
        index = ord(char) - ord('a')
        counts[index] += 1

    for char in s2:
        index = ord(char) - ord('a')
        counts[index] -= 1

    if any(counts):
        return False
        
    return True

print(CheckPermutation("abc", "cba"))
print(CheckPermutation("apple", "pale"))
```

### 計算量分析 (Complexity Analysis)

文字列の長さを $n$ とします。

| 解法 | 時間計算量 (Time) | 空間計算量 (Space) | 特徴 |
| :--- | :--- | :--- | :--- |
| **ソート法** | $O(n \log n)$ | $O(n)$ | 実装が非常に簡単。ソート処理がボトムネックとなる。 |
| **カウント法** | $O(n)$ | $O(1)$ | 文字列を1回ずつスキャンするだけで済むため、最も高速。 |

#### なぜカウント法が $O(n)$ なのか？
1. `s1` のループで $n$ 回。
2. `s2` のループで $n$ 回。
3. 合計 $2n$ 回の操作ですが、ビッグオー記法では定数を無視するため **$O(n)$** となります。

#### なぜカウント法の空間計算量が $O(1)$ なのか？
アルファベットは26文字と決まっているため、文字列がどれだけ長くなっても `counts` リストのサイズ（26）は変わりません。そのため、入力サイズに依存しない「定数空間」となります。

### 思考プロセス

1. **ソートよりカウント法を選択**
   - ソート (`sorted`) は簡潔だが $O(n \log n)$ かかる。
   - カウント法は $O(n)$ であり、大量のデータでも高速に動作する。

2. **ロジックの平坦化 (Flat Code)**
   - ループ内に複雑な `if` 文を書かず、「加算・減算・最終チェック」の3ステップに分離した。
   - これにより、コードの可読性が高まり、直感的に理解できる構造になった。

3. **Pythonの関数活用**
   - `any()` を使うことで、リスト内に残骸（0以外の数値）があるかどうかを一行で判定した。

### 結論
「計算量」と「可読性」のバランスを考え、シンプルかつ効率的なパイプライン型のロジックを構築した。
