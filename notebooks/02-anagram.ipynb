# [アルゴリズム] アナグラム判定

### 問題
2つの文字列 `s1` と `s2` が与えられたとき、一方を並べ替えてもう一方にできるか（アナグラムか）を判定してください。

**条件**:
- すべて英小文字とする。
- 実行速度を重視し、ネスト（入れ子）の少ない読みやすいコードを目指す。

```python
def CheckPermutation(s1: str, s2: str) -> bool:
    if len(s1) != len(s2):
        return False
        
    counts = [0] * 26

    for char in s1:
        index = ord(char) - ord('a')
        counts[index] += 1

    for char in s2:
        index = ord(char) - ord('a')
        counts[index] -= 1

    if any(counts):
        return False
        
    return True

print(CheckPermutation("abc", "cba"))
print(CheckPermutation("apple", "pale"))
```

### 思考プロセス

1. **ソートよりカウント法を選択**
   - ソート (`sorted`) は簡潔だが $O(n \log n)$ かかる。
   - カウント法は $O(n)$ であり、大量のデータでも高速に動作する。

2. **ロジックの平坦化 (Flat Code)**
   - ループ内に複雑な `if` 文を書かず、「加算・減算・最終チェック」の3ステップに分離した。
   - これにより、コードの可読性が高まり、直感的に理解できる構造になった。

3. **Pythonの関数活用**
   - `any()` を使うことで、リスト内に残骸（0以外の数値）があるかどうかを一行で判定した。

### 結論
「計算量」と「可読性」のバランスを考え、シンプルかつ効率的なパイプライン型のロジックを構築した。
