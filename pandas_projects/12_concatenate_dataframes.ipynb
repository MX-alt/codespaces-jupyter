## 演習：DataFrames の垂直結合（縦方向の連結）
# 【問題】
2 つの DataFrame（df1 と df2）を垂直（縦方向）に結合して、1 つの DataFrame にする。

```python
import pandas as pd

def concatenateDataFrames(df1: pd.DataFrame, df2: pd.DataFrame) -> pd.DataFrame:
    # axis=0 表示垂直拼接（按行堆叠），这也是默认值
    return pd.concat([df1, df2], axis=0)

import pandas as pd

df1 = pd.DataFrame([{"student_id": 1, "name": "Mason", "age": 8}])
df2 = pd.DataFrame([{"student_id": 2, "name": "Ava", "age": 7}])

print("--- 垂直結合の結果 ---")
print(concatenateDataFrames(df1, df2))
```

## 【まとめ】
使用関数: pd.concat([df1, df2])

重要パラメーター:

axis=0: 縦方向に結合（行を追加）。デフォルト。SQL の UNION ALL に相当。

axis=1: 横方向に結合（列を追加）。

ignore_index=True: 結合後にインデックス（行番号）を 0 から振り直す。
