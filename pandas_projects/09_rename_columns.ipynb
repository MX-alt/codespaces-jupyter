## 演習：カラム名の変更（リネーム）
# 【問題】
id, first, last, age の各カラムを、より詳細な名前に変更する。

```python
import pandas as pd

def renameColumns(students: pd.DataFrame) -> pd.DataFrame:
    # columns 参数接收一个字典：{'旧列名': '新列名'}
    return students.rename(columns={
        'id': 'student_id',
        'first': 'first_name',
        'last': 'last_name',
        'age': 'age_in_years'
    })

data = [
    [1, 'Mason', 'King', 6],
    [2, 'Ava', 'Wright', 7],
    [3, 'Taylor', 'Hall', 16]
]
df = pd.DataFrame(data, columns=['id', 'first', 'last', 'age'])

print("--- 変更前 ---")
print(df.columns.tolist())

df_renamed = renameColumns(df)

print("\n--- 変更後 ---")
print(df_renamed.columns.tolist())
```

# 【まとめ】
使用メソッド: df.rename(columns={...})

記述方法: {'旧名': '新名'} という形式の辞書（Dictionary）を引数に渡す。

メリット:

データの意味がより明確になり、可読性が向上する。

複数のカラムを一括で変更できる。
