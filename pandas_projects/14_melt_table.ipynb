## 演習：データのアンピボット（横形式から長形式への変換）
# 【問題】
四半期（quarter_1〜quarter_4）ごとに横に広がっている売上データを、縦方向の行（Q1〜Q4）に再整列（メルト）させる。

```python
import pandas as pd

def meltTable(report: pd.DataFrame) -> pd.DataFrame:
    return report.melt(
        id_vars=['product'],
        value_vars=['quarter_1', 'quarter_2', 'quarter_3', 'quarter_4'],
        var_name='quarter',
        value_name='sales'
    )

# 模拟原始的宽格式数据
data = {
    'product': ['Umbrella', 'SleepingBag'],
    'quarter_1': [417, 519],
    'quarter_2': [224, 730],
    'quarter_3': [379, 44],
    'quarter_4': [611, 394]
}
report_df = pd.DataFrame(data)

print("--- 📥 融化前（横向宽数据） ---")
print(report_df)
print("-" * 50)

# 执行融化
melted_df = meltTable(report_df)

print("--- 📤 融化后（纵向长数据） ---")
print(melted_df)
```

## 【まとめ】
使用メソッド: df.melt(id_vars=..., value_vars=..., var_name=..., value_name=...)
