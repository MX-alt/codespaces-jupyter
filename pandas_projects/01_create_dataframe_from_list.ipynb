### 1. 概要
2次元リスト student_data を、列名（student_id, age）を指定して Pandas の DataFrame に変換する。

```python
import pandas as pd

student_data = [
    [1, 15],
    [2, 11],
    [3, 11],
    [4, 20]
]

def createDataframe(student_data: list[list[int]]) -> pd.DataFrame:
    return pd.DataFrame(student_data, columns=['student_id', 'age'])

df = createDataframe(student_data)
df
```

### 2. まとめ (Simple Notes)
実装方法: pd.DataFrame(data, columns=[...]) を使うのが最も標準的。

順序: リストのインデックス順がそのまま保持される。

ポイント:

data: 元の2次元リストを渡す。

columns: 列名のリストを指定する。
