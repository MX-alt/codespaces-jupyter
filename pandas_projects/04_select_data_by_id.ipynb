【問題】
student_id が 101 の学生を抽出し、その name と age の列だけを表示する。

```python
import pandas as pd

def selectData(students: pd.DataFrame) -> pd.DataFrame:
    # 1. 筛选行：找出 student_id 等于 101 的行
    # 2. 筛选列：只取出 'name' 和 'age' 这两列
    filtered_rows = students[students['student_id'] == 101]
    return filtered_rows[['name', 'age']]

data = [
    {"student_id": 1, "name": "Mason", "age": 8},
    {"student_id": 101, "name": "Ulysses", "age": 13},
    {"student_id": 53, "name": "Eulalia", "age": 10},
    {"student_id": 128, "name": "Blaire", "age": 12}
]

students = pd.DataFrame(data)
print(selectData(students))
```

【まとめ】
行のフィルタリング: df[df['column'] == value]

内部的にはブールインデックス（True/False のリスト）が作成され、True の行だけが残る仕組み。

列の選択: df[['col1', 'col2']]

二重括弧に注意。内側の [] は列名のリストを作成し、外側の [] はそのリストを使って DataFrame から抽出するために使用する。
