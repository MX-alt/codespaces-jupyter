## 演習：条件フィルタリングとソート（並び替え）の組み合わせ
# 【問題】
体重が 100kg を超える（厳密に大きい） 動物を抽出し、体重の降順（大きい順） で名前（name）だけを返す。

```python
import pandas as pd

# 1. 你的解决方案函数
def findHeavyAnimals(animals: pd.DataFrame) -> pd.DataFrame:
    heavy_animals = animals[animals['weight'] > 100].sort_values(by='weight', ascending=False)
    return heavy_animals[['name']]

# 2. 模拟题目给出的动物园原始数据
data = {
    'name': ['Tatiana', 'Khaled', 'Alex', 'Jonathan', 'Stefan', 'Tommy'],
    'species': ['Snake', 'Giraffe', 'Leopard', 'Monkey', 'Bear', 'Panda'],
    'age': [98, 50, 6, 45, 100, 26],
    'weight': [464, 41, 106, 311, 449, 11]
}

animals_df = pd.DataFrame(data)

print("--- 原始数据（所有动物） ---")
print(animals_df)
print("-" * 50)

# 3. 调用函数筛选并排序
result_df = findHeavyAnimals(animals_df)

print("--- 🎯 过滤及降序排列后的结果 ---")
print(result_df)
```

## 【まとめ】
並び替えメソッド: df.sort_values(by='列名', ascending=False)

ascending=True: 昇順（デフォルト、小さい順）

ascending=False: 降順（大きい順）

特定の列のみ抽出: df[['列名']] （二重括弧にすることで DataFrame 形式を維持）
