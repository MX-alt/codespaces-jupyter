## 演習：データ型の変換（キャスト）
# 【問題】
浮動小数点数（float）として保存されている grade 列を 整数（int） に変換する。

```python
import pandas as pd

def changeDatatype(students: pd.DataFrame) -> pd.DataFrame:
    # 使用 astype 指定目标类型
    students['grade'] = students['grade'].astype(int)
    return students

data = [
    {"student_id": 1, "name": "Jack", "age": 13, "grade": 77.0},
    {"student_id": 2, "name": "Piper", "age": 10, "grade": 88.0}
]
df = pd.DataFrame(data)

print("--- 変換前 ---")
print(df.dtypes)  # 查看各列类型

df = changeDatatype(df)

print("\n--- 変換後 ---")
print(df.dtypes)
print(df)
```

## 【まとめ】
使用メソッド: df['列名'].astype(new_type)

主な型: int (整数), float (浮点), str (文字列/对象).

注意点:

変換時に小数点以下は切り捨てられる。

欠損値（NaN）が含まれている場合、通常の int への変換はエラーになることがある（その場合は Int64 などの nullable 型を検討する）。
