## 演習：欠損値（NaN）の含まれる行の削除
# 【問題】
name 列に欠損値（NaN）がある行をすべて削除する。

```python
import pandas as pd

def removeMissingValues(customers: pd.DataFrame) -> pd.DataFrame:
    # subset=['name'] 表示只检查 'name' 这一列是否有缺失值
    # 只要 'name' 列是 NaN，整行就会被删除
    return customers.dropna(subset=['name'])

data = [
    {"id": 1, "name": "Emma", "age": 20},
    {"id": 2, "name": None, "age": 21},    # name 缺失
    {"id": 3, "name": "Sophia", "age": 19},
    {"id": 4, "name": None, "age": 22}     # name 缺失
]

customers = pd.DataFrame(data)
print(removeMissingValues(customers))
```

## 【まとめ】
使用メソッド: df.dropna()

主要な引数:

subset=['列名']: 欠損値をチェックする特定の列を指定する（リスト形式）。指定しない場合、いずれかの列に NaN があればその行全体が削除される。
