### 演習：重複データの削除（デデュープ）
## 【問題】
email 列に基づいて重複する行を削除し、最初に現れた行だけ を残す。

```python
import pandas as pd

def dropDuplicateEmails(customers: pd.DataFrame) -> pd.DataFrame:
    # subset='email' 表示只根据 email 列来判断是否重复
    # keep='first' 表示保留第一次出现的行（这也是默认值）
    return customers.drop_duplicates(subset='email', keep='first')

data = [
    {"customer_id": 1, "name": "Ella", "email": "emily@example.com"},
    {"customer_id": 2, "name": "David", "email": "michael@example.com"},
    {"customer_id": 3, "name": "Zachary", "email": "sarah@example.com"},
    {"customer_id": 4, "name": "Alice", "email": "emily@example.com"},      # 重复 email
    {"customer_id": 5, "name": "Finn", "email": "john@example.com"},
    {"customer_id": 6, "name": "Violet", "email": "alice@example.com"}
]

customers = pd.DataFrame(data)
print(dropDuplicateEmails(customers))
```

### 【まとめ】
使用メソッド: df.drop_duplicates()

主要な引数:

subset='列名': 重複判定の基準とする列を指定。

keep='first': 最初データを残す（デフォルト）。'last' にすると最後のデータを残す。
