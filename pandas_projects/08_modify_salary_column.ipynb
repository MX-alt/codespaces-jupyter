## 演習：既存の列の更新（上書き修正）
# 【問題】
salary 列のすべての値を 2 倍 に更新（上書き）する。

```python
import pandas as pd

def modifySalaryColumn(employees: pd.DataFrame) -> pd.DataFrame:
    # 直接对原有的 'salary' 列进行重新赋值，从而覆盖旧数据
    employees['salary'] = employees['salary'] * 2
    return employees

data = [
    {"name": "Jack", "salary": 19666},
    {"name": "Piper", "salary": 74754},
    {"name": "Mia", "salary": 62509},
    {"name": "Kit", "salary": 54867}
]
df = pd.DataFrame(data)

print("--- 修正前 ---")
print(df)

print("\n--- 修正後 ---")
print(modifySalaryColumn(df))
```

## 【まとめ】
列の更新: df['既存の列名'] = df['既存の列名'] * 2 のように、同じ列名に対して再代入を行うことでデータを上書きできる。
