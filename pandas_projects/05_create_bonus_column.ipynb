### 演習：新しい列の作成（派生列）
## 【問題】
salary 列の値を 2 倍 にした新しい列 bonus を追加する。

```python
import pandas as pd

def createBonusColumn(employees: pd.DataFrame) -> pd.DataFrame:
    employees['bonus'] = employees['salary'] * 2
    return employees

# 测试代码
data = [
    {"name": "Piper", "salary": 4548},
    {"name": "Grace", "salary": 28143},
    {"name": "Georgia", "salary": 1103},
    {"name": "Willow", "salary": 6593},
    {"name": "Finn", "salary": 74576},
    {"name": "Thomas", "salary": 24433}
]
df = pd.DataFrame(data)

# 调用函数
result = createBonusColumn(df)
print(result)
```

### 【まとめ】
新機能: 列の追加と計算。

基本文法: df['新列名'] = df['既存列名'] * 係数

ベクトル演算: Pandas ではループを使わずに、列全体に対して一括で計算を行うことが可能（高速）。
