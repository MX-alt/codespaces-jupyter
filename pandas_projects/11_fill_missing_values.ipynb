## 演習：欠損値の穴埋め（フィルイン）
# 【問題】
quantity 列にある欠損値（NaN）を 0 で埋める。

```python
import pandas as pd

def fillMissingValues(products: pd.DataFrame) -> pd.DataFrame:
    # 针对 'quantity' 列，将 NaN 替换为 0
    products['quantity'] = products['quantity'].fillna(0)
    return products

data = [
    {"name": "Wristwatch", "quantity": None, "price": 135}, # quantity 缺失
    {"name": "WirelessEarbuds", "quantity": None, "price": 822}, # quantity 缺失
    {"name": "GolfClubs", "quantity": 779, "price": 9319},
    {"name": "Printer", "quantity": 849, "price": 3051}
]
df = pd.DataFrame(data)

print("--- 填充前 ---")
print(df)

df = fillMissingValues(df)

print("\n--- 填充后 ---")
print(df)
```

## 【まとめ】
使用メソッド: df['列名'].fillna(値)

使い分け:

dropna: 不完全なデータを除去する場合（第7問）。

fillna: デフォルト値（0 や "Unknown" など）で補完し、データの件数を維持する場合（今回のケース）。
