## 演習：データのピボット（長形式から横形式への変換）
# 【問題】
気象データ（weather）を回転（ピボット）させ、行を「月（month）」、列を「都市（city）」、値を「気温（temperature）」にする。

```python
import pandas as pd

def pivotTable(weather: pd.DataFrame) -> pd.DataFrame:
    # index: 决定重塑后哪一列作为“行索引”（每一行代表什么）
    # columns: 决定重塑后哪一列的“唯一值”转变为“新列名”
    # values: 决定新表格的格子里填入什么数据
    return weather.pivot(index='month', columns='city', values='temperature')

data = {
    "city": ["Jacksonville", "Jacksonville", "Jacksonville", "Jacksonville", "Jacksonville", 
             "ElPaso", "ElPaso", "ElPaso", "ElPaso", "ElPaso"],
    "month": ["January", "February", "March", "April", "May", 
              "January", "February", "March", "April", "May"],
    "temperature": [13, 23, 38, 5, 34, 20, 6, 26, 2, 43]
}

weather_df = pd.DataFrame(data)

print("--- 🔄 旋转前（原始长格式数据）---")
print(weather_df)
print("-" * 40)

# 4. 调用函数进行旋转
pivoted_df = pivotTable(weather_df)

print("--- ✨ 旋转后（优雅的宽格式数据）---")
print(pivoted_df)
```

## 【まとめ】
使用メソッド: df.pivot(index=..., columns=..., values=...)

データの形式変換:

ロング形式（縦長）: データの追加が容易で、RDB やログ収集に適した形式。

ワイド形式（横広）: 人間が見やすく、月ごとの都市間比較などがしやすい形式。
