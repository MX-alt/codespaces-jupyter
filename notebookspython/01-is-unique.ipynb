# 01. 重複のない文字列 (Is Unique)

### 問題の概要
ある文字列が、すべて異なる文字（重複なし）で構成されているかどうかを判定するアルゴリズムを実装してください。

### 制約事項 (Constraints)
* 追加のデータ構造（SetやHash Tableなど）を使用して良い場合と、使用してはいけない場合の両方を検討してください。
* 文字の種類（ASCII、Unicodeなど）によって解法が変わる可能性があることに注意してください。

## 解法1：Set(集合)

### 思考プロセス
Pythonの **`set` (集合)** を活用して解決します。集合は重複する要素を持たないという特性があるため、「元の文字列の長さ」と「重複排除後の集合の長さ」を比較することで判断できます。

### 解法の手順
1. `set(astr)` を使用して文字列を集合に変換し、重複文字を自動的に取り除きます。
2. `len(astr)`：元の文字列の長さを取得します。
3. `len(set(astr))`：重複排除後の集合の長さを取得します。
4. **ロジック判断**：両者が等しければ「重複なし」、異なれば「重複あり」と判定します。

### 計算量分析
* **時間計算量**: $O(N)$ —— 文字列を一度走査する必要があります。
* **空間計算量**: $O(N)$ —— 最悪の場合、すべての文字を保持するためのメモリが必要です。

```python
class Solution:
    def isUnique(self, astr: str) -> bool:
        return len(astr) == len(set(astr))

sol = Solution()
print(f"Testing 'abc': {sol.isUnique('abc')}")
print(f"Testing 'apple': {sol.isUnique('apple')}")
```

## 解法2：ビット演算 (Bit Manipulation)

### 思考プロセス
整数型の変数（32ビットまたは64ビット）を **ビットフラグ** として使用します。各ビットを各文字（a-z）に対応させ、出現した文字のビットを `1` に立てることで、重複を判定します。

* **メリット**: 別のデータ構造（Setなど）を使わないため、空間計算量が非常に小さい。
* **前提条件**: 文字列が `a-z` の小文字のみであると仮定（ビット範囲内に収めるため）。

### 計算量分析
* **時間計算量**: $O(N)$
* **空間計算量**: $O(1)$ —— 固定サイズの整数変数一つのみ使用。

```python
class Solution:
    def isUnique(self, astr: str) -> bool:
        mark = 0
        for char in astr:
            move_bit = ord(char) - ord('a')
            if (mark & (1 << move_bit)) != 0:
                return False
            mark |= (1 << move_bit)
        return True
    
sol = Solution()
test_str = "leetcode"
print(f"Testing '{test_str}': {sol.isUnique(test_str)}")
```

### 性能比較 (Performance Comparison)

| 解法 | 時間計算量 | 空間計算量 | 特徴 |
| :--- | :--- | :--- | :--- |
| **Set(集合)** | $O(N)$ | $O(N)$ | 実装がシンプル、可読性が高い |
| **ビット演算** | $O(N)$ | $O(1)$ | メモリ効率が最大、小文字限定 |

> **注意点**:
> このビット演算による解法は、文字が `a-z` (26文字) に限定されている場合に最適です。
> Unicodeなど全文字種に対応する必要がある場合は、**Set(集合)** による解法がより安全で汎用的です。
